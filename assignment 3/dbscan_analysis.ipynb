
"""
## Assignment 3: Clustering Algorithm Comparison

In this notebook, we compare the performance of three clustering algorithms: **DBSCAN**, **kMeans**, and **Agglomerative (Hierarchical) Clustering**. We use synthetic datasets (`make_moons`, `make_blobs`, `make_circles`) to visually assess the strengths and limitations of each algorithm.

DBSCAN is a density-based clustering method well-suited for arbitrarily shaped clusters and noise. In contrast, kMeans assumes spherical clusters, and Agglomerative clustering builds a tree-based hierarchy.

Our objective is to evaluate these methods based on cluster shape detection, noise handling, and separation.
"""


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_blobs, make_circles
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
def plot_clusters(X, labels, title):
    plt.figure(figsize=(5, 4))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
    plt.title(title)
    plt.show()
```

```python
X, _ = make_moons(n_samples=300, noise=0.05, random_state=0)
X_scaled = StandardScaler().fit_transform(X)

# DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
plot_clusters(X, dbscan_labels, "DBSCAN - Moons")

# kMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans_labels = kmeans.fit_predict(X_scaled)
plot_clusters(X, kmeans_labels, "kMeans - Moons")

# Agglomerative
agglo = AgglomerativeClustering(n_clusters=2)
agglo_labels = agglo.fit_predict(X_scaled)
plot_clusters(X, agglo_labels, "Agglomerative - Moons")
```
